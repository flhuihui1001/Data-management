## Put the IMAGE

There are two fact tables, education_fact that has the Institute names and the other foriegn keys as shown in the diagram and the other is Income Fact table that has the Agg. Gross Income and the other foriegn keys. The dimension of Year is shared by both the fact table. The location1_D has the id location_id, to which the state_zip_id of the locaiton_D references to. So we create another table which has all the states and zip combination which are not common to both the education and income dataset called spare_state_zip.

### Education fact table

```python
%%sql
DROP TABLE IF EXISTS education_fact;
CREATE TABLE education_fact (
    INS_name VARCHAR(94),
    year_id INT,
    location_id INT,
    admission_id INT,
    retention_id INT,
    sat_id INT,
    debt_id INT,
    fam_income_id INT,
    FOREIGN KEY (year_id) REFERENCES year_D(year_id),
    FOREIGN KEY (location_id) REFERENCES location_D(location_id),
    FOREIGN KEY (admission_id) REFERENCES admission_D(admission_id),
    FOREIGN KEY (retention_id) REFERENCES retention_D(retention_id),
    FOREIGN KEY (sat_id) REFERENCES sat_D(sat_id),
    FOREIGN KEY (debt_id) REFERENCES debt_D(debt_id),
    FOREIGN KEY (fam_income_id) REFERENCES fam_income_D(fam_income_id)
)
```

### Using Join to poulate the fact table

```python
%%sql
INSERT INTO education_fact 
SELECT education15.institution,year_D.year_id,location_D.location_id,admission_D.admission_id,retention_D.retention_id,sat_D.sat_id,debt_D.debt_id,fam_income_D.fam_income_id
FROM education15
JOIN
year_D
ON education15.year=year_D.year
JOIN
location_D
ON 
education15.city=location_D.city
AND education15.stabbr=location_D.state
AND education15.zip=location_D.zip
JOIN
admission_D
ON
education15.adm_rate=admission_D.adm_rate
AND education15.adm_rate_all=admission_D.adm_rate_all
JOIN
retention_D
ON
education15.ret_ft4=retention_D.ret_ft4
AND education15.ret_ftl4=retention_D.ret_ftl4_all
AND education15.ret_pt4=retention_D.ret_pt4
AND education15.ret_ptl4=retention_D.ret_ptl4
JOIN
sat_D
ON
education15.sat_avg=sat_d.sat_avg
AND education15.sat_avg_all=sat_d.sat_avg_all
JOIN
debt_d
ON
education15.debt_mdn=debt_D.debt_md
AND education15.grad_debt_mdn=debt_D.grad_debt_md
AND education15.wdraw_debt_mdn=debt_D.wdraw_debt_md
AND education15.lo_inc_debt_mdn=debt_D.lo_inc_debt_md
AND education15.md_inc_debt_mdn=debt_D.md_inc_debt_md
AND education15.hi_inc_debt_mdn=debt_D.hi_inc_debt_md
JOIN
fam_income_D
ON
education15.inc_pct_lo=fam_income_D.inc_pct_lo
AND education15.inc_pct_m1=fam_income_D.inc_pct_m1
AND education15.inc_pct_m2=fam_income_D.inc_pct_m2
AND education15.inc_pct_h1=fam_income_D.inc_pct_h1
AND education15.inc_pct_h2=fam_income_D.inc_pct_h2
```

```python
%%sql
INSERT INTO income_fact 
SELECT income.A00100,income.A00200,year_D.year_id,location1_D.location_id,income_level_D.income_level_id
FROM income
JOIN
year_D
ON income.year=year_D.year
JOIN
location1_D
ON 
income.state=location1_D.state
AND income.zip=location1_D.zip
JOIN
income_level_D
ON
income.agi_stub=income_level_D.level
```

We realised that some of the zip codes in the educaiton dataset had a - as in them. The below code updates the table by removing the -

```python
%%sql
UPDATE location_D
SET zip=t1.zip1 FROM
(SELECT city,state,location_id,split_part(zip, '-', 1) as zip1 FROM location_D)t1
where location_D.city=t1.city
AND location_D.state=t1.state
AND location_D.location_id=t1.location_id
```

### Income by State
The average income in the state of CA is very high, so it is surprising to see that the student debt is so high. One of the things that could be possible is that the Tutions are sky high in the colleges of california or that there are not a lot of local students in these colleges.
```python
%%sql
SELECT year_D.year, round(AVG(CAST(admission_D.adm_rate_all AS NUMERIC)),2) AS Admission_rate, ROUND(AVG(CAST(sat_D.sat_avg_all AS INT)),0) AS Average_SAT_score, 
    ROUND(AVG(CAST(retention_D.RET_FT4 AS NUMERIC)),2) AS Average_retention_rate, round(AVG(CAST(income_fact.A00100 AS NUMERIC)),2) AS Income, round(AVG(CAST(debt_D.debt_md AS NUMERIC)),2) AS Debt


FROM admission_D
JOIN
education_fact
ON
admission_D.admission_id=education_fact.admission_id

JOIN
sat_D
ON sat_D.sat_id=education_fact.sat_id

JOIN
retention_D
ON education_fact.retention_id=retention_D.retention_id

JOIN
year_D
ON education_fact.year_id=year_D.year_id

JOIN
location_D
ON education_fact.location_id=location_D.location_id

JOIN 
location1_D
ON location1_D.location_id=location_D.state_zip_id


JOIN
income_fact
ON location1_D.location_id=income_fact.location_id

JOIN
debt_D
ON debt_D.debt_id=education_fact.debt_id



WHERE adm_rate_all NOT LIKE 'NULL' AND RET_FT4 NOT LIKE 'NULL' AND sat_avg_all NOT LIKE 'NULL' AND debt_md NOT LIKE 'NULL' AND debt_md NOT LIKE 'PrivacySuppressed'
GROUP BY year;
```
